# 1. Web Scraping with BeautifulSoup
Use Case: Extracting data from websites for research, data analysis, or competitive intelligence.

Example: Scraping the latest news headlines from a news website.

```python
import requests
from bs4 import BeautifulSoup

# URL of the news website
url = 'https://www.bbc.com/news'

# Send a GET request to the website
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find all headline elements
headlines = soup.find_all('h3')

print("Latest News Headlines:")
for idx, headline in enumerate(headlines[:10], 1):
    print(f"{idx}. {headline.text.strip()}")
```

# 4. Password Generator
Use Case: Creating secure, random passwords for user accounts or system access.

Example: Generating a random password of user-specified length.

```python
import random
import string

def generate_password(length):
    characters = string.ascii_letters + string.digits + "!@#$%^&*()?"
    password = ''.join(random.choice(characters) for _ in range(length))
    return password

# User input for password length
try:
    length = int(input("Enter the desired password length: "))
    if length < 6:
        print("Password length should be at least 6 characters.")
    else:
        print(f"Generated Password: {generate_password(length)}")
except ValueError:
    print("Please enter a valid number.")
```

# 7. File Automation: Organizing Files into Folders
Use Case: Automatically organizing downloaded files into specific directories based on file type or other criteria.

Example: Sorting files in a directory into subfolders based on file extensions.

```python
pwd
```

```python
import os
import shutil

# Directory to organize
directory = '/content'

# Create a dictionary mapping extensions to folder names
file_types = {
    'Images': ['.jpg', '.jpeg', '.png', '.gif', '.bmp'],
    'Documents': ['.pdf', '.docx', '.txt', '.xlsx', '.pptx'],
    'Videos': ['.mp4', '.mov', '.avi', '.mkv'],
    'Music': ['.mp3', '.wav', '.aac'],
    'Archives': ['.zip', '.rar', '.tar', '.gz']
}

# Iterate over files in the directory
for filename in os.listdir(directory):
    # Skip directories
    if os.path.isdir(os.path.join(directory, filename)):
        continue

    # Get file extension
    _, ext = os.path.splitext(filename)
    ext = ext.lower()

    # Determine the target folder
    for folder, extensions in file_types.items():
        if ext in extensions:
            target_folder = os.path.join(directory, folder)
            if not os.path.exists(target_folder):
                os.makedirs(target_folder)
            shutil.move(os.path.join(directory, filename), os.path.join(target_folder, filename))
            break
    else:
        # Move unknown file types to 'Others' folder
        target_folder = os.path.join(directory, 'Others')
        if not os.path.exists(target_folder):
            os.makedirs(target_folder)
        shutil.move(os.path.join(directory, filename), os.path.join(target_folder, filename))

print("Files have been organized successfully!")
```

# 8. Building a Simple Chatbot
Use Case: Creating interactive bots for customer service, information retrieval, or entertainment.

Example: A rule-based chatbot responding to greetings.

```python
def chatbot_response(user_input):
    greetings = ['hi', 'hello', 'hey', 'greetings']
    farewell = ['bye', 'goodbye', 'see you']

    user_input = user_input.lower()

    if user_input in greetings:
        return "Hello! How can I help you today?"
    elif user_input in farewell:
        return "Goodbye! Have a great day!"
    else:
        return "I'm sorry, I don't understand. Can you please rephrase?"

# Chat loop
print("Chatbot: Hello! Type 'bye' to exit.")
while True:
    user_input = input("You: ")
    response = chatbot_response(user_input)
    print(f"Chatbot: {response}")
    if user_input.lower() in ['bye', 'goodbye', 'see you']:
        break
```

```python
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

def create_pdf(filename, text):
    c = canvas.Canvas(filename, pagesize=letter)
    width, height = letter
    c.drawString(100, height - 100, text)
    c.save()

# Create a PDF
create_pdf("hello.pdf", "Hello, this is a PDF generated by Python!")

print("PDF has been created successfully!")
```

# 10. Database Interaction with SQLAlchemy
Use Case: Interacting with databases (CRUD operations) for data management in applications.

Example: Creating a SQLite database and adding a new user.

```python
from sqlalchemy import create_engine, Column, Integer, String
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker

# Define the database
engine = create_engine('sqlite:///users.db', echo=True)
Base = declarative_base()

# Define the User model
class User(Base):
    __tablename__ = 'users'

    id = Column(Integer, primary_key=True)
    name = Column(String)
    email = Column(String)

# Create the table
Base.metadata.create_all(engine)

# Create a session
Session = sessionmaker(bind=engine)
session = Session()

# Add a new user
new_user = User(name='John Doe', email='john.doe@example.com')
session.add(new_user)
session.commit()

print("New user added to the database.")
```

# 11. Machine Learning: Classification with scikit-learn
Use Case: Building predictive models for tasks like spam detection, image recognition, or customer segmentation.

Example: Classifying iris species based on flower measurements.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Load the iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the classifier
clf = RandomForestClassifier(n_estimators=100)

# Train the model
clf.fit(X_train, y_train)

# Make predictions
predictions = clf.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test, predictions)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

# 12. Natural Language Processing: Tokenization and POS Tagging
Use Case: Analyzing and understanding the grammatical structure of text for applications like text analysis, information extraction, or chatbot development.

Example: Tokenizing a sentence and tagging parts of speech.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag

# Download required NLTK data files
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Sample sentence
sentence = "The quick brown fox jumps over the lazy dog."

# Tokenize the sentence
tokens = word_tokenize(sentence)

# Perform POS tagging
tags = pos_tag(tokens)

print("Tokens and POS Tags:")
for token, tag in tags:
    print(f"{token}: {tag}")
```

# 13. Real-Time Data Dashboard with Dash
Use Case: Creating interactive dashboards for monitoring data in real-time, useful in business analytics, operations monitoring, or data science projects.

Example: A simple dashboard displaying live updates of a random number.

```python
pip install dash
```

```python
import dash
from dash import html, dcc
from dash.dependencies import Output, Input
import random

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout
app.layout = html.Div([
    html.H1("Real-Time Random Number Dashboard"),
    html.Div(id='live-update-text'),
    dcc.Interval(
        id='interval-component',
        interval=1*1000,  # in milliseconds
        n_intervals=0
    )
])

# Define the callback to update the number
@app.callback(Output('live-update-text', 'children'),
              Input('interval-component', 'n_intervals'))
def update_number(n):
    number = random.randint(1, 100)
    return f"Current Random Number: {number}"

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

# 14. IoT Data Logging with MQTT
Use Case: Collecting and logging data from Internet of Things (IoT) devices for monitoring, analysis, or automation.

Example: Logging temperature data from an MQTT broker.

```python
pip install paho-mqtt
```

```python
import paho.mqtt.client as mqtt
import json

# MQTT settings
broker = 'broker.hivemq.com'
port = 1883
topic = 'iot/temperature'

# Callback when connected
def on_connect(client, userdata, flags, rc):
    if rc == 0:
        print("Connected to MQTT Broker!")
        client.subscribe(topic)
    else:
        print(f"Failed to connect, return code {rc}")

# Callback when a message is received
def on_message(client, userdata, msg):
    try:
        # Decode the message payload
        payload = msg.payload.decode()
        data = json.loads(payload)
        temperature = data.get('temperature')
        print(f"Temperature: {temperature}°C")

        # Here, you can add code to log to a file or database
    except json.JSONDecodeError:
        print("Received non-JSON message.")

# Initialize MQTT client
client = mqtt.Client()
client.on_connect = on_connect
client.on_message = on_message

# Connect to the broker
client.connect(broker, port, 60)

# Start the loop
client.loop_forever()
```
